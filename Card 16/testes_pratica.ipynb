Importação das bibliotecas

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input, Conv2D, Dense, Flatten, Dropout, GlobalMaxPooling2D, MaxPooling2D, BatchNormalization
from tensorflow.keras.models import Model
```

Carregamento base de dados cifar100, é uma base de dados mais complexa com 100 classes

```python
cifar100 = tf.keras.datasets.cifar100 
(x_train, y_train), (x_test, y_test) = cifar100.load_data() 
```

Normalização dos dados

```python
x_train, x_test = x_train / 255.0, x_test / 255.0 
y_train, y_test = y_train.flatten(), y_test.flatten() 

x_train.shape,  y_train.shape
```

Como aprendido em aula fiz um VGG8 para ver como reage ao CIFAR 100

```python
i = Input(shape=x_train[0].shape)

x = Conv2D(64, (3, 3), strides=2, activation='relu', padding='same')(i)
x = BatchNormalization()(x)

x = Conv2D(64, (3, 3), strides=2, activation='relu', padding='same')(x)
x = BatchNormalization()(x)
x = MaxPooling2D(2,2)(x) 

x = Conv2D(128, (3, 3),  activation='relu', padding='same')(x) 
x = BatchNormalization()(x) 

x = Conv2D(128, (3, 3), activation='relu', padding='same')(x) 
x = BatchNormalization()(x)
x = MaxPooling2D(2,2)(x) 

x = Conv2D(256, (3, 3),  activation='relu', padding='same')(x) 
x = BatchNormalization()(x) 

x = Conv2D(256, (3, 3), activation='relu', padding='same')(x) 
x = BatchNormalization()(x) 
x = MaxPooling2D(2,2)(x) 


x = Flatten()(x) 
 

x = Dense(1024, activation='relu')(x)
x = Dropout(0.2)(x) 

x = Dense(1024, activation='relu')(x)
x = Dropout(0.2)(x) 

x = Dense(100, activation='softmax')(x) 

model = Model(i, x)
```

Vou fazer 3 treinos, cada um utilizando um otimizador diferente, e ver qual se adapta melhor com essa arquitetura e esse data set

```python
model.compile(optimizer = 'adam', 
              loss = 'sparse_categorical_crossentropy', 
              metrics=['accuracy'])
r = model.fit(x_train,y_train, validation_data=(x_test, y_test), epochs = 5)
```

```python
model.compile(optimizer = 'SGD', 
              loss = 'sparse_categorical_crossentropy', 
              metrics=['accuracy'])
p = model.fit(x_train,y_train, validation_data=(x_test, y_test), epochs = 5)
```

```python
model.compile(optimizer = 'RMSprop', 
              loss = 'sparse_categorical_crossentropy', 
              metrics=['accuracy'])
l = model.fit(x_train,y_train, validation_data=(x_test, y_test), epochs = 5)
```

Após o treinamento utilizando 3 tipos de otimizadores esses são os resultados

## ADAM

```python
plt.plot(r.history['accuracy'], label='acc')
plt.plot(r.history['val_accuracy'], label='val_acc')
plt.legend()
```

```python
train_accuracy = r.history['accuracy'][-1]  
val_accuracy = r.history['val_accuracy'][-1]  
train_loss = r.history['loss'][-1]
val_loss = r.history['val_loss'][-1]

print(f"""
Métricas Finais:
- Treino:   Loss = {train_loss:.4f} | Acurácia = {train_accuracy:.4f}
- Validação: Loss = {val_loss:.4f} | Acurácia = {val_accuracy:.4f}
""")
```
