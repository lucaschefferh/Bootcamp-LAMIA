Importação das bibliotecas necessárias

```python
import pandas as pd
import tensorflow as tf
import numpy as np
import sklearn 
import scikeras
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
from tensorflow.keras.models import Sequential
from tensorflow.keras import backend as k
from tensorflow.keras import utils as np_utils
```

Carregamento dos dados

```python
base = pd.read_csv('iris_aula.csv')
x = base.iloc[:, 0:4].values  
y = base.iloc[:, 4].values  
```

Transformação para a maneira desejada da tabela Y

```python
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(y)
y = np_utils.to_categorical(y)
```

Função para criar a rede neural

```python
def criar_rede():
    k.clear_session()
    rede_neural = Sequential([
    tf.keras.layers.InputLayer(shape=(4,)),  
    tf.keras.layers.Dense(units=4, activation="relu"), 
    tf.keras.layers.Dense(units=4, activation="relu"),
    tf.keras.layers.Dense(units=3, activation="softmax") 
])
    rede_neural.compile(optimizer="Adam", loss = "categorical_crossentropy", metrics = ["categorical_accuracy"])
    return rede_neural
```

criando a rede neural com 50 epochs

```python
rede_neural = KerasClassifier(model = criar_rede, epochs=50, batch_size = 10)
```

```python
resultados = cross_val_score(estimator=rede_neural, X=x,y=y, cv=10, scoring="accuracy")
```
