Criação de um dicionário para repor os valores faltantes com os respectivos valores mais aparentes de cada categoria.

```python
valores = {'vehicleType' : 'limousine',
           'gearbox' : 'manuell',
           'model' : 'golf',
           'fuelType' : 'benzin',
           'notRepairedDamage' : 'nein'}
```

```python
bd = bd.fillna(value = valores)
```

Criação dos dados de teste retirando o valor

```python
X = bd.iloc[:, 1:12].values
X
```

Tabela com os valores para poder fazer a verificação

```python
y = bd.iloc[:, 0].values
y
```

Encodificação das colunas necessárias

```python
onehotencoder = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(), [0, 1, 3, 5, 8, 9, 10])], remainder='passthrough')
```

```python
X = onehotencoder.fit_transform(X).toarray()
X
```

```python
bd.shape, X.shape
```

Criação da rede neural com 316 entradas, 2 camadas ocultas com 158 neuronios, e uma camada de saída com apenas 1.

```python
regressor = Sequential([
    tf.keras.layers.InputLayer(shape =(316,)),
    tf.keras.layers.Dense(units = 158, activation = 'relu'),
    tf.keras.layers.Dense(units = 158, activation = 'relu'),
    tf.keras.layers.Dense(units = 1, activation = 'linear'),
])
regressor.summary()
```

passagem dos parametros para a rede neural

```python
regressor.compile(loss = 'mean_absolute_error', optimizer = 'adam', metrics = ['mean_absolute_error'])
```

treinamento

```python
regressor.fit(X, y, batch_size = 300, epochs = 100)
```

```python
previsoes = regressor.predict(X)
```

Previsões ao lado dos preços verdadeiros

```python
previsoes, y
```

Média de ambos

```python
previsoes.mean(), y.mean()
```

Salvamento para possível futuro uso.
