```python
import pandas as pd
import numpy as np
import tensorflow as tf
import sklearn
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
from tensorflow.keras import utils as np_utils
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
base = pd.read_csv('iris_aula.csv')
```

```python
base
```

```python
X = base.iloc[:, :4].values
X
```

```python
y = base.iloc[:, 4].values
y
```

#### setosa -> 0
#### versicolor -> 1
#### virginica -> 2
                 

```python
label_encoder = LabelEncoder()
y =  label_encoder.fit_transform(y)
y
```

Transformação em uma forma que a rede neural consiga reproduzir, e transformação em um array de 150 exemplos e 3 categorias

```python
y = np_utils.to_categorical(y)
y
```

Criação dos dados de teste e treinamento

```python
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size = 0.25)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

Criação da rede neural

```python
rede_neural = Sequential([
    tf.keras.layers.InputLayer(shape = (4,)),
    tf.keras.layers.Dense(units = 4, activation = 'relu'),
    tf.keras.layers.Dense(units = 4, activation = 'relu'),
    tf.keras.layers.Dense(units = 3, activation = 'softmax')    
])
```

```python
rede_neural.summary()
```

```python
rede_neural.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['categorical_accuracy'])
```

```python
rede_neural.fit(X_train, y_train, batch_size = 10, epochs = 500)
```

No treinamento ele teve um accuracy de 98% e um loss de 3%, já na base de test ele teve um accuracy de 95% e um loss de 5%

```python
rede_neural.evaluate(X_test, y_test)
```

```python
previsoes = rede_neural.predict(X_test)
print(previsoes)
```

```python
previsoes = previsoes > 0.5
print(previsoes)
```

Transformamos os dados de teste em suas respectivas iris, como ja foi passada.

```python
y_test2 = [np.argmax(t) for t in y_test]
print(y_test2)
```

Transformamos os dados de previsoes em suas respectivas iris, como ja foi passada.

```python
previsoes2 = [np.argmax(t) for t in previsoes]
previsoes2
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test2, previsoes2)
```

```python
confusion_matrix(y_test2, previsoes2)
```
